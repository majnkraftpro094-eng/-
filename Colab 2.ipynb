```python
from transformers import pipeline

# Завантажуємо моделі
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
generator = pipeline("text-generation", model="distilgpt2")

def ai_support_bot(user_review):
    # Перетворюємо відгук у нижній регістр для легшого пошуку слів
    review_lower = user_review.lower()

    # 1. Нова умова: Тригер на ціну (Завдання з Кроку 3)
    if "price" in review_lower or "expensive" in review_lower:
        start_phrase = "We always try to keep our prices fair. Let us explain..."
    else:
        # 2. Аналіз тональності (якщо про ціну не йдеться)
        analysis = classifier(user_review)[0]
        sentiment = analysis['label']

        # 3. Зміна стилю (Твоє завдання: нові фрази)
        if sentiment == "POSITIVE":
            start_phrase = "Huge thanks for your kind words! We're glad that"
        else:
            start_phrase = "We're sorry to hear you're unhappy. Regarding"

    # 4. Генерація тексту (temperature=1.5 для "хаосу")
    response = generator(f"{start_phrase} {user_review[:40]}",
                         max_new_tokens=30,
                         temperature=1.5,  # Експеримент з хаосом
                         do_sample=True)

    # Визначаємо, що повернути для виводу
    final_sentiment = "PRICE_ISSUE" if "price" in review_lower else classifier(user_review)[0]['label']
    return final_sentiment, response[0]['generated_text']

# Перевірка (спробуй змінити текст на "The price is too high")
my_review = "The price is a bit high, but the quality is okay!"
label, reply = ai_support_bot(my_review)

print(f"Аналіз: {label}")
print(f"Відповідь AI: {reply}")
```
